## Transformation 

```python
#merging company tweet, and tweet dfs on "tweet_id"
merged_df = pd.merge(tweet_id, tweet_df, on="tweet_id")
merged_df.head()
```

```python
merged_df.ticker_symbol.unique()
```

```python
# Since we have 2 google, removing all the googl ticker 
merged_df = merged_df[merged_df.ticker_symbol !="GOOGL"]
```

```python
merged_df.ticker_symbol.unique()
```

```python
#converting unix epoch to date format similar
merged_df["post_date"] = pd.to_datetime(merged_df["post_date"], unit='s').dt.strftime('%Y-%m-%d')
merged_df.sample(20)
```

```python
#checking values before dropping null values
merged_df.count()
```

```python
#dropping null values
merged_df = merged_df.dropna()
```

```python
#checking values/count after dropping null values
merged_df.count()
```

```python
# Checking start and end date of the tweet data 
merged_df = merged_df.sort_values(by='post_date')
merged_df.head()
```

```python
merged_df.tail()
```

```python
# Double check nulls
merged_df.isnull().sum()
```

```python
# remane post_date 
merged_df = merged_df.rename(columns={'post_date': 'Date'}, errors='raise')
merged_df.head()
```

Apple


```python
df_aapl_f = df_aapl
df_aapl_f
```

```python
#df_aapl_f['Date'] = pd.to_datetime(df_aapl_f.index)
#df_aapl_f["Date"] = pd.to_datetime(df_aapl_f["Date"], unit='s').dt.strftime('%Y-%m-%d')

df_aapl_f = df_aapl_f.reset_index()
df_aapl_f["Date"] = pd.to_datetime(df_aapl_f["Date"], unit='s').dt.strftime('%Y-%m-%d')
df_aapl_f
```

```python
final_apple['price_action'] = final_apple['Close'] - final_apple['Open']

final_apple.tail()
```

```python
final_apple = final_apple.drop(columns=['Dividends','Stock Splits'])
final_apple
```

```python
final_apple["Ticker"] = "AAPL"
final_apple['Percent_Change'] = ((final_apple["Close"]-final_apple["Open"])/final_apple["Open"])*100
final_apple
```

```python
final_apple.to_csv('final_aapl.csv', index=False)
```
